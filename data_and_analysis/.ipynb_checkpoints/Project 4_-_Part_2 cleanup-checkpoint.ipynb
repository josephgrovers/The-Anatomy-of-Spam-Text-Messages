## Readying data for model

```python
def untokenize(text):
    # re-gather word lists back into messages!
    text = ' '.join(text)
    return text
```

```python
def tokenize(text):
    # split text by spaces / tokenize
    text = text.split(' ')
    return text
```

```python
data['text_untokenized'] = data['text'].apply(lambda x : untokenize(x))
```

```python
# Grab all the texts
texts = data['text_untokenized']
```

```python
# Create a list of strings, one for each text
texts_list = [text for text in texts]
```

```python
# Collapse the list of strings into a single long string for processing
big_text_string = ' '.join(texts_list)
```

```python
words_list = tokenize(big_text_string)
```

## Averaging word embeddings

```python
def document_vector(word2vec_model, doc):
    # remove out-of-vocabulary words
    doc = [word for word in doc if word in model.wv.vocab]
    return np.mean(model[doc], axis=0)

# Our earlier preprocessing was done when we were dealing only with word vectors
# Here, we need each document to remain a document 
def preprocess(text):
    text = text.lower()
    doc = word_tokenize(text)
    # doc = [word for word in doc if word not in stopwords] # couldn't get this working
    doc = [word for word in doc]
    doc = [word for word in doc if word.isalpha()] 
    return doc

# Function that will help us drop documents that have no word vectors in word2vec
def has_vector_representation(word2vec_model, doc):
    """check if at least one word of the document is in the
    word2vec dictionary"""
    return not all(word not in word2vec_model.wv.vocab for word in doc)

# Filter out documents
def filter_docs(corpus, texts, condition_on_doc):
    """
    Filter corpus and texts given the function condition_on_doc which takes a doc. The document doc is kept if condition_on_doc(doc) is true.
    """
    number_of_docs = len(corpus)

    if texts is not None:
        texts = [text for (text, doc) in zip(texts, corpus)
                 if condition_on_doc(doc)]

    corpus = [doc for doc in corpus if condition_on_doc(doc)]

    print("{} docs removed".format(number_of_docs - len(corpus)))

    return (corpus, texts)
```

```python
len(texts_list)
```

```python
# Preprocess the corpus
from nltk.corpus import stopwords
corpus = [preprocess(text) for text in texts_list]

# # leaving this out, messes with ordering of rows for spam/ham labels for later. This only removes 22 documents anyway
# Remove docs that don't include any words in W2V's vocab
corpus, texts_list = filter_docs(corpus, texts_list, lambda doc: has_vector_representation(model, doc))
# Filter out any empty docs
corpus, texts_list = filter_docs(corpus, texts_list, lambda doc: (len(doc) != 0))

x = []

# use this part to make a cleaned x, but we're preserving all
for doc in corpus: # append the vector for each document
    x.append(document_vector(model, doc))
    
X = np.array(x) # list to array
```

### Time to actually clean

```python
def clean_text(text):
    # remove symbols
    text = re.sub('[^a-zA-Z]', ' ', text)  
    # make lowercase
    text = text.lower()  
    
    # split text by spaces / tokenize
    text = text.split(' ')
    # remove stop words from word lists
    text = [w for w in text if not w in set(stopwords.words('english'))] 
    # remove empty strings in lists
    text = [w for w in text if w != ""]
    # remove 'lt' strings in lists
    text = [w for w in text if w != "lt"]
    # remove 'gt' strings in lists
    text = [w for w in text if w != "gt"]
    # re-gather word lists back into messages! (For Word2Vec, we want to keep it tokenized)
    # text = ' '.join(text)
    
    return text
```

```python
# This takes ~17 seconds
data['text'] = data['text'].apply(lambda x : clean_text(x))
```

### Trying to build a visualization based off of stack exchange post

```python
w2v_vectors = model.wv.vectors # here you load vectors for each word in your model
w2v_indices = {word: model.wv.vocab[word].index for word in model.wv.vocab} # here you load indices - with whom you can find an index of the particular word in your model
```

```python
def vectorize(line): 
    words = []
    for word in line: # line - iterable, for example list of tokens 
        try:
            w2v_idx = w2v_indices[word]
        except KeyError: # if you does not have a vector for this word in your w2v model, continue 
            continue
        words.append(w2v_vectors[w2v_idx])
        if words: 
            words = np.asarray(words)
            min_vec = words.min(axis=0)
            max_vec = words.max(axis=0)
            return np.concatenate((min_vec, max_vec))
        if not words:
            return None 
```

## Word Clouds

I'll untokenize and retokenize in this section since the wordclouds 

```python
def untokenize(text):
    # re-gather word lists back into messages!
    text = ' '.join(text)
    return text
```

```python
data['text'] = data['text'].apply(lambda x : untokenize(x))
```

```python
# checking the most common words in spam messages
spam = ' '.join(text for text in data['text'][data['label'] == 1])

wordcloud = WordCloud(background_color = 'gray', max_words = 50, height = 1000, width = 1000).generate(spam)

plt.axis('off')
plt.title('Most Common Words in Spam Messages', fontsize = 20)
plt.imshow(wordcloud);
```

```python
ham = ' '.join(text for text in data['text'][data['label'] == 0])

wordcloud = WordCloud(background_color = 'gray', max_words = 50, height = 1000, width = 1000).generate(ham)

plt.axis('off')
plt.title('Most Common Words in Ham Messages', fontsize = 20)
plt.imshow(wordcloud);
```
